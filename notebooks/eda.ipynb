## Dataset Overview

The dataset contains the following columns:

1. **TransactionId**: Unique transaction identifier on the platform.
2. **BatchId**: Unique number assigned to a batch of transactions for processing.
3. **AccountId**: Unique number identifying the customer on the platform.
4. **SubscriptionId**: Unique number identifying the customer subscription.
5. **CustomerId**: Unique identifier attached to the account.
6. **CurrencyCode**: Country currency.
7. **CountryCode**: Numerical geographical code of the country.
8. **ProviderId**: Source provider of the item bought.
9. **ProductId**: Item name being bought.
10. **ProductCategory**: Product IDs are organized into these broader product categories.
11. **ChannelId**: Identifies if the customer used web, Android, iOS, pay later, or checkout.
12. **Amount**: Value of the transaction. Positive for debits from the customer account and negative for credit into the customer account.
13. **Value**: Absolute value of the amount.
14. **TransactionStartTime**: Transaction start time.
15. **PricingStrategy**: Category of Xente's pricing structure for merchants.
16. **FraudResult**: Fraud status of the transaction (1 = yes, 0 = no).

## EDA Tasks Performed

The `EDA` class performs the following tasks:

### 1. **Dataset Overview**
   - Displays the number of rows and columns in the dataset.
   - Shows the data types of each column.
   - Displays the first 5 rows of the dataset.

### 2. **Summary Statistics**
   - Calculates and displays summary statistics for all columns, including:
     - Count, mean, standard deviation, min, max, and quartiles for numerical columns.
     - Count, unique values, and mode for categorical columns.

### 3. **Numerical Feature Distribution**
   - Visualizes the distribution of numerical columns (`Amount` and `Value`) using histograms.
   - Helps identify patterns, skewness, and potential outliers in the data.

### 4. **Categorical Feature Distribution**
   - Analyzes the distribution of categorical columns (`CurrencyCode`, `CountryCode`, `ProductCategory`, `ChannelId`, and `FraudResult`) using bar plots.
   - Provides insights into the frequency and variability of categories.

### 5. **Correlation Analysis**
   - Computes the correlation matrix for numerical columns (`Amount` and `Value`).
   - Visualizes the correlation matrix using a heatmap to understand relationships between numerical features.

### 6. **Missing Values**
   - Identifies and displays missing values in the dataset.
   - Logs a warning if missing values are found.

### 7. **Outlier Detection**
   - Detects outliers in numerical columns (`Amount` and `Value`) using box plots.
   - Helps identify extreme values that may affect the analysis.

### 8. **Logging**
   - Logs the progress of each task to both a file (`eda.log`) and the console.
   - Logs warnings for missing values and other potential issues.

## How to Use the Code

1. **Load the Dataset**:
   - Replace `'data.csv'` with the path to your dataset.
   - Instantiate the `EDA` class:
     ```python
     eda = EDA('data.csv')
     ```

2. **Run All EDA Tasks**:
   - Call the `run_all()` method to perform all EDA tasks sequentially:
     ```python
     eda.run_all()
     ```

3. **View Results**:
   - The results of each task will be displayed in the notebook.
   - Logs will be saved to `eda.log` and printed to the console.

# Outlier Detection Analysis: Box Plots of `Amount` and `Value`

This section analyzes the **box plots** for the `Amount` and `Value` features from the dataset. The box plots help us understand the distribution of these features, identify outliers, and determine if the data is skewed.

---

## Box Plot of `Amount`

### Observations:
1. **Outliers**: 
   - The box plot for `Amount` shows several outliers beyond the upper whisker. These outliers represent transactions with unusually high amounts compared to the majority of the data.
   
2. **Spread**: 
   - The interquartile range (IQR, the box itself) is relatively narrow, indicating that most transaction amounts are concentrated within a small range.

3. **Skewness**: 
   - The presence of outliers on the higher side suggests that the `Amount` distribution is **right-skewed**. This means there are a few transactions with significantly higher amounts pulling the distribution to the right.

---

## Box Plot of `Value`

### Observations:
1. **Outliers**: 
   - Similar to `Amount`, the `Value` box plot also shows outliers beyond the upper whisker. These outliers represent transactions with unusually high absolute values.

2. **Spread**: 
   - The IQR for `Value` is also narrow, indicating that most transaction values are concentrated within a small range.

3. **Skewness**: 
   - The presence of outliers on the higher side suggests that the `Value` distribution is also **right-skewed**. This means there are a few transactions with significantly higher values pulling the distribution to the right.

---

## General Observations

1. **Similarity in Distributions**:
   - Both `Amount` and `Value` exhibit similar patterns in their distributions, with narrow IQRs and right-skewed tails due to outliers.

2. **Impact of Outliers**:
   - The outliers in both `Amount` and `Value` could significantly impact statistical analyses or machine learning models. For example:
     - They may inflate the mean of the distributions.
     - They could affect the performance of models sensitive to extreme values (e.g., linear regression).

3. **Data Quality**:
   - The presence of outliers may indicate data quality issues, such as errors in data entry or unusual transactions that need further investigation.

---

## Recommendations

1. **Outlier Handling**:
   - Investigate the outliers to determine if they are genuine transactions or errors.
   - If they are genuine, consider using robust statistical methods (e.g., median instead of mean) or models that are less sensitive to outliers (e.g., tree-based models).
   - If they are errors, remove or correct them.

2. **Data Transformation**:
   - Apply transformations like **log transformation** or **winsorization** to reduce the impact of outliers.

3. **Further Analysis**:
   - Analyze the outliers to understand their characteristics (e.g., are they from a specific product category, channel, or customer segment?).
